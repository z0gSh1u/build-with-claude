# 01j - 流式响应

```python
from pprint import pprint
from dotenv import load_dotenv
load_dotenv()
import os
# 检查 API Key 和 Base URL 是否已配置
bool(os.environ["ANTHROPIC_API_KEY"] and os.environ["BASE_URL"])
```

```python
from anthropic import Anthropic

# 初始化 Anthropic 客户端
client = Anthropic(api_key=os.environ["ANTHROPIC_API_KEY"], base_url=os.environ["BASE_URL"])
# 本系列教程将使用 Claude Sonnet 4 模型
model = "claude-sonnet-4-20250514"
```

```python
final_message = None

with client.messages.stream(
    model=model,
    max_tokens=1000,
    messages=[{
        "role": "user",
        "content": "Describe what's relational database, in about 50 words."
    }]
) as stream:
    for text in stream.text_stream:
        print(text, end="")

    final_message = stream.get_final_message()
```

```python
pprint(final_message.to_dict())
```
